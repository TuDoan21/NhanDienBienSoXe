```python
yaml_content = """
train: /content/drive/MyDrive/DoAn_AI/data_bienso/dataset/train/images
val: /content/drive/MyDrive/DoAn_AI/data_bienso/dataset/valid/images
test: /content/drive/MyDrive/DoAn_AI/data_bienso/dataset/test/images

nc: 1
names: ['bien_so']
"""

with open('/content/drive/MyDrive/DoAn_AI/data_bienso/dataset/data.yaml', 'w') as f:
    f.write(yaml_content)
print("Đã cập nhật file data.yaml với đường dẫn mới thành công!")
```

```python
from ultralytics import RTDETR

model = RTDETR('rtdetr-l.pt')

results = model.train(
    data='/content/drive/MyDrive/DoAn_AI/data_bienso/dataset/data.yaml',
    epochs=25,
    imgsz=640,
    batch=8,
    workers=2,
    project='/content/drive/MyDrive/DoAn_AI/KetQua',
    name='train_rtdetr_1'
)
```
